```python
import pandas as pd

df = pd.read_csv("patients.csv")
df
```

```python
df['first_name'] = df['first_name'].fillna("Unknown")
df
```

```python
df.loc[:, 'last_name'] = df['last_name'].str.rstrip("#").str.title()
df
```

```python
import numpy as np

df.loc[(df['age'] < 0) | (df['age'] > 120), 'age'] = None
df
```

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='mean')
df[['age']] = imputer.fit_transform(df[['age']])
df
```

```python
df['age'] = df['age'].astype(int)
```

```python
df['gender'] = df['gender'].str.strip().str.upper()
df['gender'] = df['gender'].replace({'M': 'Male', 'F': 'Female'})
df
```

```python
df = df.drop_duplicates(subset=['patient_id'])
df
```

```python
print("Cleaned dataset:\n", df.head())
```

```python
cleaned_patients = "patients_cleaned.csv"
df.to_csv(cleaned_patients, index=False)
```
