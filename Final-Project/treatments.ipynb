```python
import pandas as pd

df = pd.read_csv("treatments.csv")
df
```

```python
df = df.drop_duplicates().copy()
df
```

```python
from sklearn.impute import SimpleImputer

doc_imputer = SimpleImputer(strategy="most_frequent")
df.loc[:, "doctor_name"] = doc_imputer.fit_transform(df[["doctor_name"]])
df
```

```python
cost_imputer = SimpleImputer(strategy="median")
df.loc[:, "treatment_cost"] = cost_imputer.fit_transform(df[["treatment_cost"]])
df
```

```python
df.loc[df["doctor_name"].str.contains("\?\?\?", na=False), "doctor_name"] = "Unknown Doctor"
df.loc[df["department"] == "UnknownDept", "department"] = "Unknown Department"
df
```

```python
df.loc[:, "doctor_name"] = df["doctor_name"].str.strip().str.title()
df.loc[:, "department"] = df["department"].str.strip().str.title()
df
```

```python
df.loc[:, "admission_date"] = pd.to_datetime(df["admission_date"], errors="coerce")
df
```

```python
import numpy as np

df.loc[df["treatment_cost"] < 0, "treatment_cost"] = np.nan
df.loc[:, "treatment_cost"] = cost_imputer.fit_transform(df[["treatment_cost"]])
df
```

```python
df.loc[df["treatment_cost"] > 200000, "treatment_cost"] = 200000
df
```

```python
df = df.astype({
    "admission_id": "int",
    "patient_id": "int",
    "diagnosis_id": "int",
    "department": "str",
    "doctor_name": "str"
})
```

```python
print("Cleaned Treatments Data:\n", df.head(10))
```

```python
cleaned_treatments = "treatments_cleaned.csv"
df.to_csv(cleaned_treatments, index=False)
```
