```python
# Imports
# Import create_engine: A core function from SQLAlchemy used to create a connection "engine" 
# (a common interface to the database).
from sqlalchemy import create_engine
# Import pandas: Used for data analysis and, in this context, likely for 
# executing queries and loading results into a DataFrame (e.g., pd.read_sql).
import pandas as pd
# Import matplotlib.pyplot: The foundational library for plotting in Python.
import matplotlib.pyplot as plt
# Import seaborn: A high-level visualization library based on matplotlib, 
# commonly used for statistical graphics.
import seaborn as sns

# Database Connection
# Define the database connection string and create the connection engine.
# Format: "dialect+driver://username:password@host:port/database_name"
# - "mysql+pymysql": Specifies the 'mysql' dialect using the 'pymysql' driver.
# - "root:@localhost": Connects as user 'root' with an empty password (indicated by ':')
#   to the 'localhost' server.
# - "healthcare_dw": The specific database (data warehouse) to connect to.
engine = create_engine("mysql+pymysql://root:@localhost/healthcare_dw")

# Test Connection (Optional but Recommended)
# Use a 'with' block to create a connection from the engine.
# This pattern is a best practice as it automatically handles opening
# and (most importantly) closing the connection, even if errors occur.
with engine.connect() as conn:
    # If this print statement executes, the connection was successful.
    print("Connected to MySQL successfully!")
```

```python
# Define the SQL Query
# Define a SQL query to aggregate data from the data warehouse.
# This query calculates the average treatment cost and total case count
# for each department, joining the fact and dimension tables.
query = """
SELECT 
    d.department_name AS department,
    -- Round the average cost to 2 decimal places for clean reporting.
    ROUND(AVG(f.treatment_cost), 2) AS avg_treatment_cost,
    -- Count the number of admissions (rows) for each group.
    COUNT(*) AS total_cases
FROM fact_admissions f
-- Join the fact table (transactions) with the department dimension table (attributes).
JOIN dim_department d ON f.department_id = d.department_id
-- Group the results by the department's name to perform the aggregations.
GROUP BY d.department_name
-- Order the final results from highest to lowest average cost.
ORDER BY avg_treatment_cost DESC;
"""

# Execute Query and Load Data
# Use pandas to execute the SQL query against the pre-configured 'engine'.
# pd.read_sql handles opening a connection, running the query, and loading
# the resulting dataset directly into a DataFrame.
df_cost = pd.read_sql(query, engine)

# Create Visualization
# Create a new figure (the canvas) with a specific size (8x5 inches)
# for better readability.
plt.figure(figsize=(8,5))

# Generate a bar plot using seaborn.
# - data=df_cost: Specifies the DataFrame to use.
# - x="department": Maps the 'department' column to the x-axis.
# - y="avg_treatment_cost": Maps the 'avg_treatment_cost' column to the y-axis.
# - palette="pastel": Uses a color-friendly, light-colored palette.
# - hue="department": Assigns a different color to each department bar.
# - legend=False: Disables the legend, as the x-axis labels are sufficient.
sns.barplot(data=df_cost, x="department", y="avg_treatment_cost", palette="pastel", hue="department", legend=False)

# Format and Save the Plot
# Set the chart title and axis labels for clarity.
plt.title("Average Treatment Cost per Department")
plt.xlabel("Department")
plt.ylabel("Average Treatment Cost")

# Rotate the x-axis labels by 45 degrees to prevent them
# from overlapping if they are long.
plt.xticks(rotation=45)

# Adjust plot parameters for a tight layout, ensuring labels don't get cut off.
plt.tight_layout()

# Save the generated plot as a high-resolution PNG file.
# - dpi=300: Sets dots-per-inch to 300, a standard for high-quality images.
# - bbox_inches="tight": Ensures the saved file includes all plot elements.
plt.savefig("avg_treatment_cost.png", dpi=300, bbox_inches="tight")

# Display the plot in the script's output window (e.g., in a Jupyter Notebook).
plt.show()
```

```python
# Define the SQL Query
# Define a SQL query to track disease cases over time (by month).
query = """
SELECT 
    dd.diagnosis_desc AS disease,
    DATE_FORMAT(dt.full_date, '%%Y-%%m') AS month,
    COUNT(*) AS cases
FROM fact_admissions f
JOIN dim_diagnosis dd ON f.diagnosis_id = dd.diagnosis_id
JOIN dim_date dt ON f.date_id = dt.date_id
GROUP BY dd.diagnosis_desc, month
ORDER BY month;
"""

# Execute Query and Load Data
# Execute the query using the existing 'engine' and load the results
# directly into a new pandas DataFrame called 'df_disease'.
df_disease = pd.read_sql(query, engine)

# Create Visualization
# Create a new figure (canvas) with a 10x6 inch size for a wider view,
# which is good for time-series plots.
plt.figure(figsize=(10,6))

# Generate a line plot using seaborn.
# - x="month": The time component is on the x-axis.
# - y="cases": The metric (count) is on the y-axis.
# - hue="disease": This is key. It tells seaborn to draw a *separate line*
#   for each unique value in the 'disease' column, automatically
#   creating a legend.
# - marker="o": Adds a small circle marker at each data point for clarity.
sns.lineplot(data=df_disease, x="month", y="cases", hue="disease", marker="o")

# Format and Save the Plot
# Set the chart title and axis labels.
plt.title("Disease Trends Over Time")
plt.xlabel("Month")
plt.ylabel("Number of Cases")

# Rotate the x-axis labels (months) for better readability,
# especially if there are many months.
plt.xticks(rotation=45)

# Adjust plot layout to prevent labels from being cut off.
plt.tight_layout()

# Save the plot as a high-resolution PNG file for reports.
plt.savefig("disease_trend.png", dpi=300, bbox_inches="tight")

# Display the plot in the output window.
plt.show()
```

```python
# Define the SQL Query
# Define a SQL query to calculate key performance indicators (KPIs) for each doctor.
query = """
SELECT 
    d.doctor_name,
    -- Count unique patient IDs to get the distinct number of patients each doctor saw.
    COUNT(DISTINCT f.patient_id) AS total_patients,
    -- Count all admissions to get the total number of treatments performed.
    COUNT(f.admission_id) AS total_treatments,
    -- Sum all treatment costs to get the total revenue generated by each doctor.
    ROUND(SUM(f.treatment_cost), 2) AS total_revenue,
    -- Calculate the average cost for a single treatment by each doctor.
    ROUND(AVG(f.treatment_cost), 2) AS avg_cost_per_treatment
FROM 
    fact_admissions f
-- Join with the doctor dimension to get the doctor's name from their ID.
JOIN dim_doctor d ON f.doctor_id = d.doctor_id
-- Group all calculations by the doctor's name.
GROUP BY 
    d.doctor_name
ORDER BY 
    -- Sort the results from the most treatments to the least.
    total_treatments DESC;
"""

# Execute Query and Load Data
# Execute the SQL query against the database engine.
# The results are loaded directly into a pandas DataFrame named 'df_doctor'.
df_doctor = pd.read_sql(query, engine)

# Create Visualization
# Set up a new matplotlib figure with a specific size (8x5 inches).
plt.figure(figsize=(8,5))

# Create a bar plot using seaborn to visualize the total revenue.
# - x="doctor_name": Maps the doctor's name to the x-axis.
# - y="total_revenue": Maps the calculated total revenue to the y-axis.
# - palette="muted": Uses a color-blind-friendly, muted color palette.
# - hue="doctor_name": Assigns a different color to each doctor's bar.
# - legend=False: Disables the legend, as the x-axis labels are sufficient.
sns.barplot(data=df_doctor, x="doctor_name", y="total_revenue", palette="muted", hue="doctor_name", legend=False)

# Format and Save the Plot
# Set the chart title and axis labels for clarity.
plt.title("Doctor Performance (Total Revenue)")
plt.xlabel("Doctor")
plt.ylabel("Total Revenue")

# Apply a tight layout to ensure all elements (like labels) fit neatly.
plt.tight_layout()

# Save the plot as a high-resolution PNG image.
plt.savefig("doctor_performance.png", dpi=300, bbox_inches="tight")

# Display the plot in the output window.
plt.show()
```
